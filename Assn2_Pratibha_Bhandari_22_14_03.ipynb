````python
##Drop all rows with N/A,NA,na values in Num_Bedroom
print(H['NUM_BEDROOMS'].unique())
H=H.replace('na',np.nan)
H
````

````python
H.dropna(subset=['NUM_BEDROOMS'],inplace=True,axis=0)
H
````

````python
##to correct the indexing after dropping some rows
H.index=range(0,6)
H
````

````python
#Replace the NaN and String entries in Num_Bath with previous entry
print(H['NUM_BATH'].unique())
for i in range(0,6):
    if ((H['NUM_BATH'][i] is np.nan) or (H['NUM_BATH'][i] == 'HURLEY')):
        H.replace(H['NUM_BATH'][i],H['NUM_BATH'][i-1], inplace=True)
H
````

````python
##Replace the empty values in “SQ_FT” with the mean of the all entries.
print(H['SQ_FT'].unique())
H.replace("--", np.nan, inplace=True)
H
````

````python
H.info()
````

````python
#to convert data type of SQ_FT as float as it is wrongly classified as float in the dataframe
print(H['SQ_FT'].values)
H['SQ_FT']=H['SQ_FT'].astype('float64')
H.info()
````

````python
#replacing the nan values with mean of the column SQ_FT
H['SQ_FT']=H['SQ_FT'].fillna(H['SQ_FT'].mean())
H
````

````python
##the sum of SQ_FT of all entries in the dataset
H['SQ_FT'].values.sum()
````

````python
#Add a new column “NUM_STORE” and with all the values as 1 for all rows
H['NUM_STORE'] = 1
````
